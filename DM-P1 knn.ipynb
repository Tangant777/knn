```python
#imports
import pandas as pd
import numpy as np
```

```python
#read data file
Rice=pd.read_csv(r'/content/Rice.csv')
Rice
```

```python
#replace unknown sells with null
Rice.replace('???',np.nan,inplace=True)
```

```python
#change wrong data types to float
Rice['Area']=Rice['Area'].astype('float64')
Rice['Minor_Axis_Length']=Rice['Minor_Axis_Length'].astype('float64')
Rice['Convex_Area']=Rice['Convex_Area'].astype('float64')
Rice.dtypes
```

```python
#Number of rows and colomns
print("Number of rows: ",len(Rice))
print("Number of columns: ",len(Rice.columns))
```

```python
#min, Max of features
print("min of features:\n",Rice.min())
print("\n\nMax of features:\n",Rice.max())
```

```python
#found count of nulls
Rice.isnull().sum()
```

```python
#delete nulls
Rice.dropna(inplace=True)
Rice
```

```python
#found count of duplicates
Rice.duplicated().sum()
```

```python
#delete duplicates
Rice.drop_duplicates(inplace=True)
Rice
```

```python
#feature means
Rice.mean()
```

```python
#feature stds
Rice.std()
```

```python
#drop labels for Computing
Rice2=Rice.drop(['Class'],axis=1)

#normalize
Rice2=(Rice2-Rice2.mean())/Rice2.std()
Rice2
```

```python
#check mean and std to be true
print(Rice2.mean())
print(Rice2.std())
```

```python
#get new data from user
new_data=[0,0,0,0,0,0,0]
print("input your data in this order: Area Perimeter Major_Axis_Length	Minor_Axis_Length	Eccentricity	Convex_Area	Extent")
for i in range(7):
  new_data[i]=float(input())
new_data
```

```python
#normalize new data
new_data=(new_data-Rice.mean())/Rice.std()
new_data
```

```python
#calculate new data and primitive data distance
#in new column "distance"
Rice2['distance']=np.zeros(3800)
for i in range(7):
  Rice2['distance']+=abs(Rice2-new_data).iloc[:, i]
Rice2['distance']
```

```python
#return "class" column to dataframe
Rice2['Class']=Rice['Class']
Rice2
```

```python
#return copy df to primary df
Rice=Rice2
```

```python
#sort datas relative to distance
Rice=Rice.sort_values('distance')
Rice
```
